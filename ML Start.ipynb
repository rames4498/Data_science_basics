```python
def find_outliers(x):
    q1=x.quantile(.25)
    q3=x.quantile(.75)
    iqr = q3-q1
    floor = q1 - 1.5*iqr
    ceiling = q3+1.5*iqr
    outlier_indices = list(x.index[(x<floor)|(x>ceiling)])
    outlier_values = list(x[outlier_indices])
    return outlier_indices,outlier_values
```

```python
a = find_outliers(dataf.Age)
age_indices,age_values = find_outliers(dataf['Age'])
print("Outliers for Age")
print(np.sort(age_values))
```
